```python
def custom_sigmoid(Z):
        exponents = np.exp(-Z)
        sigmoids = 1 / (1 + exponents)
        return sigmoids

```

```python
def custom_softmax(probability):
        probability[probability >= 0.7] = 1
        probability[probability < 0.7] = 0
        y_predict = probability.astype(int)
        return y_predict

```

```python
def predict(X, weights, bias):
        Z = np.dot(X, weights) + bias
        probabilities = custom_sigmoid(Z)
        y_predict = custom_softmax(probabilities)
        return y_predict

```

```python
def compute_cost(y_predict, y_actual, n_examples):
        epsilon = 1e-5
        cost = np.sum(
            (y_actual * np.log(y_predict+epsilon)) + ((1 - y_actual) * np.log(1 - y_predict+epsilon))
        )
        cost = -(1 /n_examples) * cost
        return cost
```

```python
def compute_grad(y_predict, y_actual, X, n_examples):
        d_activations = y_predict - y_actual
        d_weights = (1 /n_examples) * (np.dot(X.T, d_activations))
        d_bias = (1 / n_examples) * np.sum(d_activations)
        grads = {"d_weights": d_weights, "d_bias": d_bias}
        return grads
```

```python
def update_params(weights, bias, alpha, grads):
    weights = weights - (alpha * grads["d_weights"])
    bias = bias - (alpha * grads["d_bias"])
    return weights, bias
```

```python
weights = np.random.randn(n_features, 1)
bias = 0
alpha = 0.003
y_pred = predict(X, weights, bias)
cost = compute_cost(y_pred, y, n_examples)

def run_iterations(X, y_pred, y , weights, bias, alpha, n_examples):
    y = y
    y_pred = y_pred
    X = X
    weights = weights
    bias = bias 
    alpha = alpha
    n_examples = n_examples
    for i in range(1, 100):
        y_pred = predict(X, weights, bias)
        cost = compute_cost(y_pred, y, n_examples)
        grads = compute_grad(y_pred, y, X, n_examples)
        weights, bias = update_params(weights, bias, alpha, grads)
        print(f"cost in iteration {i} is: {cost}")

```
